```python
import os
import json
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# =====================================================================
# 1. PATH CONFIGURATION & INIT
# =====================================================================
print("🥇 Starting TalentMatch AI Production Engine...")
final_data_path = './challenge_data/[PUB] India_runs_data_and_ai_challenge/India_runs_data_and_ai_challenge'
candidates_file = f"{final_data_path}/candidates.jsonl"

# Load lightweight, high-performance semantic text model
print("🤖 Loading Embedding Transformer Model (all-MiniLM-L6-v2)...")
model = SentenceTransformer('all-MiniLM-L6-v2')

# =====================================================================
# 2. LOAD & PRE-FILTER DATA (Stage 1: Scaling down 100,000 records)
# =====================================================================
print("📦 Loading 100,000 candidates from JSONL (this may take a moment)...")
df = pd.read_json(candidates_file, lines=True)

print("⚡ Stage 1 Filter: Flattening features safely and filtering target pool...")

def safe_extract_list(items):
    """Safely extracts strings from lists, even if nested as dicts."""
    if not isinstance(items, list):
        return ""
    extracted = []
    for item in items:
        if isinstance(item, str):
            extracted.append(item)
        elif isinstance(item, dict):
            # Grab the first string value found in the dictionary (e.g., 'name', 'value')
            str_vals = [str(v) for v in item.values() if isinstance(v, str)]
            if str_vals:
                extracted.append(str_vals[0])
    return ", ".join(extracted)

def clean_profile_text(row):
    skills = safe_extract_list(row.get('skills'))
    certifications = safe_extract_list(row.get('certifications'))

    headline = ""
    if isinstance(row.get('profile'), dict):
        headline = row['profile'].get('headline', '')

    return f"Role: {headline}. Skills: {skills}. Certifications: {certifications}."

# Apply our new bulletproof string parser
df['combined_text'] = df.apply(clean_profile_text, axis=1)

# Grab a high-potential subset of 2,000 records with keyword indicators to run fast embeddings
high_potential_df = df[df['combined_text'].str.contains('ML|AI|Machine Learning|Deep Learning|Data Scientist|Data Science', case=False, na=False)].copy()
high_potential_df = high_potential_df.head(2000).reset_index(drop=True)
print(f"✅ Filtered down to {len(high_potential_df)} high-potential candidates for deep vector matching.")

# =====================================================================
# 3. DEEP SEMANTIC VECTOR MATCHING (Stage 2)
# =====================================================================
target_jd_intent = """
Applied ML/AI Engineer Data Scientist. 6-8 years total experience, 4-5 years in product companies.
Shipped end-to-end ranking, search, or recommendation systems to real users at meaningful scale.
Expertise in retrieval frameworks, hybrid vs dense search, evaluation loops, and LLM fine-tuning integration.
Located in or willing to relocate to Noida or Pune. High engagement and activity signals.
"""

print("🧠 Generating contextual vector embeddings...")
jd_vector = model.encode([target_jd_intent])
candidate_vectors = model.encode(high_potential_df['combined_text'].tolist(), show_progress_bar=True)

print("⚖️ Computing multi-signal hybrid matrix scores...")
semantic_scores = cosine_similarity(candidate_vectors, jd_vector).flatten()
high_potential_df['semantic_score'] = semantic_scores

def extract_active_signal(signals):
    if isinstance(signals, dict):
        # Look for typical activity/engagement signal key names
        for key in ['activity_score', 'platform_activity', 'score', 'engagement']:
            if key in signals:
                try: return float(signals[key])
                except: pass
    return 50.0

high_potential_df['activity_signal'] = high_potential_df['redrob_signals'].apply(extract_active_signal)
high_potential_df['normalized_activity'] = high_potential_df['activity_signal'] / 100.0

# 70% Semantic context match, 30% Live activity engagement indicator
high_potential_df['score'] = (high_potential_df['semantic_score'] * 0.7) + (high_potential_df['normalized_activity'] * 0.3)

# Sort strictly by descending scores
high_potential_df = high_potential_df.sort_values(by='score', ascending=False).reset_index(drop=True)

# =====================================================================
# 4. GENERATING DYNAMIC REASONING & OUTPUT FORMATTING
# =====================================================================
print("✍️ Generating contextual reasoning logs for top tier talent...")

def create_reasoning(row):
    headline = "AI Specialist"
    if isinstance(row.get('profile'), dict):
        headline = row['profile'].get('headline', 'AI Specialist')
    return f"{headline} displaying strong contextual alignment. Context text match score ({row['semantic_score']:.2f}) bolstered by positive active behavioral engagement signals on the platform."

final_top_100 = high_potential_df.head(100).copy()
final_top_100['rank'] = range(1, 101)
final_top_100['reasoning'] = final_top_100.apply(create_reasoning, axis=1)

# Precise column output formatting required by rules
submission_output = final_top_100[['candidate_id', 'rank', 'score', 'reasoning']]

output_csv_path = 'team_talentmatch_ai_submission.csv'
submission_output.to_csv(output_csv_path, index=False)

print("\n" + "="*60)
print("🏆 WINNING SUBMISSION DATA GENERATED COMFORTABLY! 🏆")
print("="*60)
print(f"Saved exactly 100 rows to: '{output_csv_path}'")
print("\nTop 5 Ranked Standouts Peek:")
print(submission_output.head(5).to_string(index=False))
```
